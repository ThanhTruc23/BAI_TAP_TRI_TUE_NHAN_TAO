2. Các thành phần của Agent theo kiến trúc Model-Based
State (Trạng thái hiện tại):
Là trạng thái hiện tại của bàn cờ tại thời điểm agent đang xử lý.
- State bao gồm vị trí của tất cả các ô số và ô trống trong ma trận 3x3.
- Percept:
Là dữ liệu đầu vào mà agent nhận được từ môi trường sau mỗi lần thực hiện hành động.
Percept trong bài toán 8-puzzle chính là ma trận hiện tại của trò chơi, giúp agent xác định:
vị trí ô trống
các nước đi hợp lệ
trạng thái hiện tại đã đạt goal hay chưa
- Model:
Hiểu biết về môi trường:
Agent hiểu được cách môi trường thay đổi khi thực hiện hành động.
Cụ thể:
ô trống chỉ được di chuyển trong phạm vi ma trận
ô trống có thể đổi chỗ với ô liền kề theo 4 hướng
mỗi lần di chuyển sẽ tạo ra trạng thái mới
một số trạng thái có thể bị lặp lại nên cần được kiểm tra
Lưu trữ thông tin (Internal State):
Agent lưu trữ:
trạng thái đích (Goal State)
danh sách trạng thái đã duyệt
Queue phục vụ cho thuật toán BFS
đường đi từ trạng thái ban đầu đến trạng thái hiện tại
kế hoạch di chuyển tối ưu để đạt Goal
- Rules (Tập luật và Lập luận IF):
IF trạng thái hiện tại bằng Goal State THEN Action = "STOP"
IF trạng thái mới chưa xuất hiện trong danh sách visited THEN thêm trạng thái vào Queue để tiếp tục tìm kiếm
IF ô trống có thể di chuyển theo hướng hợp lệ THEN sinh trạng thái con tương ứng
IF Queue vẫn còn phần tử THEN tiếp tục mở rộng node tiếp theo
IF Queue rỗng THEN kết thúc chương trình vì không tìm thấy lời giải
3. Trạng thái môi trường (GOAL STATE)

Trạng thái đích đạt được khi tất cả các ô số được sắp xếp đúng thứ tự tăng dần từ trái sang phải, từ trên xuống dưới và ô trống nằm ở vị trí cuối cùng.
Goal State
1 2 3
4 5 6
7 8 0

```python

# SIMPLE 8-PUZZLE 
# Trạng thái ban đầu
start = [
    [1, 2, 3],
    [4, 0, 6],
    [7, 5, 8]
]

# Goal State
goal = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 0]
]
# In ma trận
def print_board(board):

    for row in board:
        print(row)

    print()
# Tìm vị trí ô trống
def find_zero(board):

    for i in range(3):
        for j in range(3):

            if board[i][j] == 0:
                return i, j
# trạng thái mới

def generate(board):

    states = []

    x, y = find_zero(board)

    # UP
    if x > 0:

        new_board = [row[:] for row in board]

        new_board[x][y], new_board[x-1][y] = new_board[x-1][y], new_board[x][y]

        states.append(new_board)

    # DOWN
    if x < 2:

        new_board = [row[:] for row in board]

        new_board[x][y], new_board[x+1][y] = new_board[x+1][y], new_board[x][y]

        states.append(new_board)

    # LEFT
    if y > 0:

        new_board = [row[:] for row in board]

        new_board[x][y], new_board[x][y-1] = new_board[x][y-1], new_board[x][y]

        states.append(new_board)

    # RIGHT
    if y < 2:

        new_board = [row[:] for row in board]

        new_board[x][y], new_board[x][y+1] = new_board[x][y+1], new_board[x][y]

        states.append(new_board)

    return states
# BFS

queue = [start]

visited = []

found = False

while len(queue) > 0:

    current = queue.pop(0)

    print("Trạng thái hiện tại:")
    print_board(current)

    # Goal Test
    if current == goal:

        print("Đã tìm thấy Goal State")

        found = True

        break

    visited.append(current)

    # Sinh trạng thái mới
    next_states = generate(current)

    for state in next_states:

        if state not in visited and state not in queue:

            queue.append(state)

# KẾT QUẢ
if found:

    print("Chương trình kết thúc")

else:

    print("Không tìm thấy lời giải")
```
